```python
import pandas as pd #import libraries
import numpy as np
```

```python
station_df=pd.read_csv(r'location_station.csv',sep=';') #load dataset
lat_lon_station_df=pd.read_csv(r'lat_lon_station.csv',sep=';')
```

```python
lat_lon_station_df
```

```python
station_df
```

```python
merge_station_df=pd.merge(station_df,lat_lon_station_df,on='Station',how='inner') #join 2 dataset into 1
```

```python
merge_station_df
```

```python
filter_station = merge_station_df[merge_station_df['Province'].isin(['JawaBarat', 'Banten', 'JawaTengah','JawaTimur'])]  #filter dataset 
```

```python
filter_station=filter_station.sort_values(by='Province').reset_index(drop=True) #sorting dataset
```

```python
#merge_station_df =merge_station_df.drop(columns=['Altitude (m)']) #drop irrelevant column
```

```python
filter_station["Latitude"]=filter_station["Latitude"].str.replace(',','.') #change the characters
filter_station["Longitude"]=filter_station["Longitude"].str.replace(',','.')
```

```python
filter_station["Latitude"]=filter_station["Latitude"].astype(float) #change datatype
filter_station["Longitude"]=filter_station["Longitude"].astype(float)
```

```python
filter_station.sort_values(by='Province') #show the dataframes
```

```python
min_longitude = 110.5
max_longitude = 113.5
filter_station = filter_station[(filter_station['Longitude'] >= min_longitude) & (filter_station['Longitude'] <= max_longitude)].reset_index(drop=True)
```
